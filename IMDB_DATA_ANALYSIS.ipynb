# About The Dataset

1. imdb_id:it representss the imdb_id of the movie in the IMDB Database.
2. popularity:it represents the popularity score of movie.
3. budget:it represents the budget of the movie in dollers.
4. revenue:it represents the total revenue collected by the movie in dollers

# Data Wrangling

```python
df.head(10)
```

```python
df.info()
```

```python
# droping duplicates
df.drop_duplicates(inplace=True)
df.info()
```

```python
# keeping only those columns which are needed further
md = df[['popularity','budget','revenue','original_title','runtime','genres','vote_count','vote_average','release_year']]
```

```python
# creating new variable profit
md['profit'] = md['revenue'] - md['budget']
```

```python
md.info()
```

```python
md.describe()
```

```python
md.isnull().sum()
```

```python
md.dropna(subset=['genres'],inplace=True)
# it will only remove the null values from the genres
```

```python
md.isnull().sum()
```

```python
md.head()
```

```python
s = md['genres'].str.split('|').apply(pd.Series, 1).stack()
s.index = s.index.droplevel(-1)
s.name = 'genres'
del md['genres']
md = md.join(s)
```

```python
md.head()
```

```python
md.shape
```

```python
corr = md.drop(['original_title','genres'], axis=1, inplace=False).corr()
sns.heatmap(corr, annot=True,cmap='BuPu')
```

# Q1.Which genre are the most common (number of movies made)?

```python
num_movies_per_genre = md_genre.original_title.nunique().sort_values(ascending=False)
print(num_movies_per_genre)
```

```python
plt.figure(figsize=(8,8))
plt.pie(x=num_movies_per_genre.values,labels=num_movies_per_genre.index,autopct='%1.1f%%')
plt.title('movies_per_genre in %')
plt.show()
```

# Which genre have high avg. budget and revenue ?

```python
sorted_by_budget = md_genre_budget.sort_values(by='budget',ascending=False)
md_genre_revenue = pd.DataFrame(md_genre['revenue'].mean()).reset_index()
sorted_by_revenue = md_genre_revenue.sort_values(by='revenue',ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x="genres",y="budget",data=sorted_by_budget)
plt.title("average movie budget by genre (highest first)")
plt.xlabel("genre")
plt.ylabel("average budget")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(x="genres",y="revenue",data=sorted_by_revenue)
plt.title("average movie revenue by genre (highest first)")
plt.xlabel("genre")
plt.ylabel("average budget")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# in the graphs we clearly see that the genre adventure has both the highest avg. budget and revenue.Fantasy comes in budget and revenue.
